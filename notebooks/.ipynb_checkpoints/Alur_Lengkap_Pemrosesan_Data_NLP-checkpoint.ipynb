## 2. Alur Preprocessing & OCR (Hanya Mistral OCR)
File terkait: `services/ocr_service.py`, `services/ocr_preprocessor.py`

### 2.1 Preprocessing Gambar (Opsional untuk Optimisasi)
1. Gambar dibaca menggunakan OpenCV (jika tersedia)
2. Gambar di-resize: **max dimensi 1800px** (untuk mengurangi payload dan waktu proses)
3. Gambar di-re-encode menjadi **JPG dengan kualitas 82%** (untuk kompresi ringkas tapi tetap jelas)
4. Hasil preprocessing disimpan ke file temporary baru

### 2.2 Encode Gambar ke Base64
1. Gambar (yang sudah di-preprocessing atau asli) dibaca dalam mode binary
2. Gambar di-encode menjadi string Base64 untuk dikirim ke Mistral API
3. Format URL: `data:image/jpeg;base64,<base64-string>`

### 2.3 Kirim ke Mistral OCR API
1. Sistem membuat **HTTP POST Request** ke endpoint: `https://api.mistral.ai/v1/chat/completions`
2. **Header Request**:
   - `Authorization`: `Bearer <MISTRAL_API_KEY>` (dari environment variable)
   - `Content-Type`: `application/json`
3. **Payload Request**:
   ```json
   {
     "model": "pixtral-large-latest",
     "messages": [
       {
         "role": "user",
         "content": [
           {"type": "text", "text": "Extract semua teks dari gambar, hanya kembalikan teksnya tanpa penjelasan"},
           {"type": "image_url", "image_url": "data:image/jpeg;base64,<base64-string>"}
         ]
       }
     ],
     "temperature": 0,
     "max_tokens": 2048
   }
   ```
4. **Timeout**:
   - Connect: 30 detik
   - Read: 180 detik
   - Max Attempt: 3x percobaan (jika gagal, akan coba lagi dengan jeda)

---

### Tahap 4.1: Koreksi Teks OCR (Normalisasi)
Fungsi: `koreksi_teks()` di `nlp/normalizer.py`

**Step-by-step koreksi teks**:

1. **Pass 0 (Preprocess)**: Menghapus angka `0` atau `1` yang menempel di akhir kata (contoh: `prmen1` → `prmen`, `will0` → `willo`)
2. **Pass 1**: Semua teks diubah menjadi **lowercase** (untuk konsistensi processing)
3. **Pass 2**: Mengganti angka `0` → `o` dan `1` → `l` di tengah kata (contoh: `will0` → `willo`, `w1llo` → `wllo`)
4. **Lindungi Frasa Status**: Menyembunyikan frasa status multi-kata (misal: `belum lunas`, `sudah bayar`)
5. **Pass 1.5 (Prioritas Kamus Alias)**: Cek apakah kata ada di `KAMUS_ALIAS`
6. **Pass 2**: Replace typo dan singkatan dari `NORMALIZATION_DICT`
7. **Pass 3 (Fuzzy Matching)**: Koreksi typo dengan `rapidfuzz` (Bigram: 90%, Unigram: 85%)
8. **Restore Frasa Status**

````python
# Contoh Penggunaan NLP Pipeline
from nlp.processor import proses_nlp
from core.master_data import get_master_data, get_context_data

# Dapatkan master data
master_data, _ = get_context_data()
db_metode = get_master_data()
daftar_barang = db_metode['barang'] if db_metode else []
mapping_metode = {m['nama']: m['keyword'] for m in db_metode['metode']} if db_metode else {}

# Contoh teks OCR (misalnya hasil dari gambar budis.jpg)
contoh_teks_ocr = """
Nota Penjualan
Tanggal: 29-06-2025
Kepada: Pak Budi
Barang: 2 dus Willo
Harga: Rp 100.000
Total: Rp 200.000
Metode: Tunai Lunas
""".strip()

print("=== Hasil OCR yang akan diproses ===")
print(contoh_teks_ocr)

print("\n=== Hasil Proses NLP ===")
nlp_result = proses_nlp(
    teks_user=contoh_teks_ocr,
    db_metode=db_metode,
    daftar_barang=daftar_barang,
    mapping_metode=mapping_metode
)

for i, r in enumerate(nlp_result['valid_results']):
    print(f"\nTransaksi #{i+1}")
    for key, value in r.items():
        print(f"{key}: {value}")
````
