# Panduan Lengkap: Alur Pemrosesan Data NLP Bot Penjualan

Notebook ini menunjukkan secara detail bagaimana data input diproses dari awal hingga menjadi data transaksi yang terstruktur dan siap disimpan ke database.

Semua kode dan logika yang ditampilkan **murni diambil dari modul sistem yang sesungguhnya**, bukan simulasi.

---

## Alur Kerja Sistem:

### **A. Preprocessing & Normalisasi** (`nlp/normalizer.py`)
1. Input Data → `handle_text_message()` di `handlers/text_handler.py`
2. Sanitasi Input → `sanitize_input()` di `utils/security.py`
3. Pemisahan Multi-Entry → `split_multi_entries()` di `nlp/processor.py`
4. Text Cleaning & Normalisasi (Multi-Pass):
   - Pass 0: Hapus angka menempel di huruf (regex)
   - Pass 1: Lowercase
   - Pass 2-3: Ganti angka 0→o, 1→l di dalam kata (regex)
   - Pass 1.5: Pencocokan Eksak pada `KAMUS_ALIAS` (dictionary lookup)
   - Pass 2: Koreksi Typo & Singkatan dari `NORMALIZATION_DICT` (dictionary lookup)
   - Pass 3: Fuzzy Matching Bigram/Unigram menggunakan **RapidFuzz** (`fuzz.ratio`)

### **B. Ekstraksi Entitas** (`nlp/extractor.py`)
5. Slang Preprocessor → Normalisasi singkatan gaul (regex)
6. Fuzzy Intent Fallback → `fuzzy_intent_fallback()` di `nlp/intent_matcher.py` (RapidFuzz)
7. Ekstraksi Entitas (Regex Rule-Based):
   - TANGGAL (relatif & absolut)
   - NAMA (dengan prefix gelar & fallback)
   - AKSI / Intent (keyword matching)
   - BARANG & SATUAN (Master Data + KAMUS_ALIAS)
   - JUMLAH, HARGA, TOTAL (regex angka)
   - STATUS & METODE_PEMBAYARAN (keyword matching)

### **C. Identifikasi Intent** (`nlp/extractor.py` + `nlp/intent_matcher.py`)
8. `tentukan_intent()` → Menentukan intent final (Regex + RapidFuzz `match_intent_from_dataset`)

### **D. Post-Processing** (`nlp/processor.py`)
9. `_apply_generic_command_guards()` → Penyesuaian AKSI untuk perintah CRUD
10. `_apply_multi_overrides()` → Context Inheritance & Override Status/Metode

### **E. Perhitungan Data Transaksi** (`handlers/text_handler.py`)
11. Lookup Harga dari Master Data → `cari_harga_default()`
12. Perhitungan Total → `Total = Jumlah × Harga Satuan`
13. Logika Setengah/Cicilan → `NOMINAL_BAYAR = Total ÷ 2`
14. Fallback Default Values (tanggal hari ini, status dari nominal)

### **F. Penyusunan Data Transaksi** (`handlers/text_handler.py`)
15. Validasi Field Wajib → `[TANGGAL, NAMA, BARANG, JUMLAH, TOTAL, STATUS, METODE_PEMBAYARAN]`
16. Simpan ke Session → `user_sessions[chat_id]`
17. Tampilkan Resume → `susun_balasan_resume()` di `services/ui_transaksi.py`

## **Langkah A2: Pemisahan Multi-Entry**

Fungsi: `split_multi_entries()` di `nlp/processor.py` (baris 8-58)

Memisahkan teks yang berisi lebih dari satu transaksi menggunakan:

## **Langkah A3: Preprocessing Multi-Pass (`koreksi_teks`)**

Fungsi: `koreksi_teks()` di `nlp/normalizer.py` (baris 10-174)

Ini adalah inti preprocessing. Sistemnya menggunakan **6 pass berurutan**:


````python
print("="*70)
print("DETAIL PREPROCESSING MULTI-PASS")
print("="*70)

entri_contoh = entries[0]
print(f"\n📥 INPUT: '{entri_contoh}'")

# === PASS 0: Hapus angka menempel di huruf (regex) ===
print("\n--- PASS 0: Hapus angka menempel (regex) ---")
teks_p0 = re.sub(r'(\b[a-zA-Z]+)([01])\b', r'\1', entri_contoh, flags=re.IGNORECASE)
print(f"  Regex: re.sub(r'(\\b[a-zA-Z]+)([01])\\b', r'\\1', teks)")
print(f"  Hasil: '{teks_p0}'")
print(f"  Contoh: 'prmen1' → 'prmen' (angka 1 dihapus)")

# === PASS 1: Lowercase ===
print("\n--- PASS 1: Lowercase ---")
teks_p1 = teks_p0.lower()
print(f"  Hasil: '{teks_p1}'")

# === PASS 2-3: Ganti angka 0→o, 1→l di dalam kata (regex) ===
print("\n--- PASS 2-3: Ganti 0→o, 1→l di dalam kata (regex) ---")
teks_p2 = re.sub(r'([a-zA-Z])0([a-zA-Z]*)', r'\1o\2', teks_p1)
teks_p2 = re.sub(r'([a-zA-Z])1([a-zA-Z]*)', r'\1l\2', teks_p2)
print(f"  Regex: re.sub(r'([a-zA-Z])0([a-zA-Z]*)', r'\\1o\\2', teks)")
print(f"  Hasil: '{teks_p2}'")
print(f"  Contoh: 'will0' → 'willo' (0 diganti o)")

# === PASS 1.5: Pencocokan Eksak KAMUS_ALIAS ===
print("\n--- PASS 1.5: Pencocokan Eksak KAMUS_ALIAS (dictionary lookup) ---")
words_pre = teks_p2.split()
replaced = []
for w in words_pre:
    match = re.match(r'^([^\w]*)(\w+)([^\w]*)$', w)
    if match:
        prefix, word_part, suffix = match.groups()
        w_low = word_part.lower()
        if w_low in KAMUS_ALIAS:
            print(f"  ✅ '{w_low}' ditemukan di KAMUS_ALIAS → '{KAMUS_ALIAS[w_low]}'")
            replaced.append(f"{prefix}{KAMUS_ALIAS[w_low]}{suffix}")
        else:
            replaced.append(w)
    else:
        replaced.append(w)
teks_p3 = " ".join(replaced)
print(f"  Hasil: '{teks_p3}'")
````

````python
# === PASS 2 (Lanjutan): Typo & Singkatan dari NORMALIZATION_DICT ===
print("--- PASS 2 (Lanjutan): Koreksi Typo & Singkatan (NORMALIZATION_DICT) ---")
_typo_map = NORMALIZATION_DICT.get("typo_corrections", {})
_abbr_map = NORMALIZATION_DICT.get("abbreviations", {})
print(f"  Jumlah typo_corrections: {len(_typo_map)}")
print(f"  Jumlah abbreviations: {len(_abbr_map)}")
print(f"  Contoh typo_corrections: {dict(list(_typo_map.items())[:3])}")
print(f"  Contoh abbreviations: {dict(list(_abbr_map.items())[:3])}")

# === PASS 3: Fuzzy Matching (RapidFuzz) ===
print("\n--- PASS 3: Fuzzy Matching Bigram/Unigram (RapidFuzz) ---")
print("  Library: rapidfuzz (process.extractOne + fuzz.ratio)")
print("  Threshold Bigram: ≥ 90")
print("  Threshold Unigram: ≥ 85")
print("  Threshold Status: ≥ 80")

# Demonstrasi fuzzy matching
contoh_typo = ["prmen", "mses", "lunwas"]
keywords_target = DAFTAR_KATA_KUNCI
status_keywords = {"lunas", "hutang", "belum", "sudah", "setengah"}

for typo in contoh_typo:
    # Cek status keywords dulu
    status_res = process.extractOne(typo, list(status_keywords), scorer=fuzz.ratio)
    if status_res and status_res[1] >= 80:
        print(f"  '{typo}' → status match: '{status_res[0]}' (skor: {status_res[1]}%)")
        continue
    
    # Cek unigram
    res = process.extractOne(typo, keywords_target, scorer=fuzz.ratio)
    if res:
        terbaik, skor = res[0], res[1]
        status = "✅ MATCH" if skor >= 85 else "❌ NO MATCH"
        mapped = KAMUS_ALIAS.get(terbaik, terbaik)
        print(f"  '{typo}' → '{terbaik}' (skor: {skor}%) {status}")
        if skor >= 85:
            print(f"           → KAMUS_ALIAS: '{mapped}'")
````

````python
# Hasil akhir koreksi_teks (fungsi lengkap dari sistem)
print("\n" + "="*70)
print("HASIL KOREKSI TEKS LENGKAP (koreksi_teks dari normalizer.py)")
print("="*70)

for i, entry in enumerate(entries, 1):
    teks_dikoreksi = koreksi_teks(entry)
    
    print(f"\nEntri {i}:")
    print(f"  Sebelum: '{entry}'")
    print(f"  Sesudah: '{teks_dikoreksi}'")
````

# **B. EKSTRAKSI ENTITAS**

Modul utama: `nlp/extractor.py` → fungsi `ekstrak_entitas()`

Teknik yang digunakan:
- **Regex** (`re.search`) untuk mendeteksi pola tanggal, jumlah, harga, nama
- **Dictionary Lookup** pada `KAMUS_ALIAS` untuk mencocokkan nama barang
- **RapidFuzz** (`fuzzy_intent_fallback`) untuk mencocokkan teks dengan dataset training

### Tahapan di dalam `ekstrak_entitas()`:
1. **Slang Preprocessor** (baris 126-208): Normalisasi singkatan menggunakan `slang_map` + regex
2. **Fuzzy Intent Fallback** (baris 216): `fuzzy_intent_fallback()` mencari contoh paling mirip di dataset
3. **Ekstraksi TANGGAL** (baris 293-377): Regex untuk tanggal relatif & absolut
4. **Deteksi AKSI** (baris 379-449): Keyword matching untuk menentukan jenis perintah
5. **Deteksi NAMA** (baris 451-551): Regex dengan prefix (pak/bu/mas) + fallback tanpa prefix
6. **Deteksi BARANG** (baris 553-620): Master Data lookup → KAMUS_ALIAS fallback
7. **Deteksi JUMLAH & SATUAN** (baris 622-682): Regex pattern matching
8. **Deteksi HARGA & TOTAL** (baris 684-694): Regex + `parse_price_shorthand()`
9. **Deteksi STATUS** (baris 707-724): Keyword matching prioritas (Dicicil > Hutang > Lunas)
10. **Deteksi METODE_PEMBAYARAN** (baris 726-758): Mapping dinamis + fallback statik


````python
print("="*70)
print("EKSTRAKSI ENTITAS (ekstrak_entitas dari extractor.py)")
print("="*70)

for i, entry in enumerate(entries, 1):
    teks_koreksi = koreksi_teks(entry)
    hasil_ekstraksi = ekstrak_entitas(teks_koreksi, teks_asli=entry)
    
    print(f"\n--- ENTITAS ENTRI {i} ---")
    print(f"Teks asli   : '{entry}'")
    print(f"Teks koreksi: '{teks_koreksi}'")
    
    entitas = hasil_ekstraksi['entitas']
    for key, value in entitas.items():
        if value not in (None, False, []):
            print(f"  {key:20s} : {value}")
````

# **C. IDENTIFIKASI INTENT**

Modul: `nlp/extractor.py` → `tentukan_intent()` (baris 973-1040) + `nlp/intent_matcher.py`

Teknik: **Dua lapis** pencocokan:
1. **RapidFuzz** (`match_intent_from_dataset`): Fuzzy match input terhadap semua pattern di dataset training menggunakan `fuzz.token_set_ratio` (threshold ≥ 80)
2. **Rule-Based Fallback**: Jika fuzzy tidak match atau skor rendah, gunakan mapping manual berdasarkan nilai `AKSI` dan `KONTEKS_AGREGASI`


````python
print("="*70)
print("IDENTIFIKASI INTENT")
print("="*70)

# Demonstrasi match_intent_from_dataset (RapidFuzz)
contoh_inputs = [
    "pak andi pesan permen 10 dus lunas tunai",
    "tampilkan transaksi hari ini",
    "cek harga produk",
    "bayar hutang pak rudi 100000",
]

print("\n--- Fuzzy Intent Matching (match_intent_from_dataset) ---")
print("Library: rapidfuzz (fuzz.token_set_ratio)")
print("Threshold: ≥ 80\n")

for inp in contoh_inputs:
    intent, score = match_intent_from_dataset(inp.lower())
    print(f"  Input: '{inp}'")
    print(f"  → Intent: {intent or 'None'} (skor: {score}%)")
    print()

# Demonstrasi tentukan_intent (rule-based)
print("--- Penentuan Intent Final (tentukan_intent) ---")
for i, entry in enumerate(entries, 1):
    teks_koreksi = koreksi_teks(entry)
    hasil = ekstrak_entitas(teks_koreksi, teks_asli=entry)
    entitas = hasil['entitas']
    intent = tentukan_intent(entitas, entry.lower())
    print(f"  Entri {i}: intent = '{intent}' (AKSI: {entitas.get('AKSI')})")
````

# **D. POST-PROCESSING**

Modul: `nlp/processor.py`

### D1: `_apply_generic_command_guards()` (baris 648-711)
Menyesuaikan AKSI untuk perintah CRUD Master Data menggunakan **regex** (`re.search`):
- "cek harga barang" → `Cek Harga Barang`
- "tambah barang" → `Tambah Barang`
- "hapus barang" → `Hapus Barang`
- "update status" → `Update Status`

### D2: `_apply_multi_overrides()` (baris 135-393)
Menangani kasus multi-entry:
- **Context Inheritance**: Salin NAMA dan TANGGAL dari entri pertama ke entri lain yang kosong
- **Global Override**: "semua lunas" → terapkan ke semua item
- **Per-Item Override**: "bembeng sudah lunas tf" → terapkan hanya ke item bembeng
- **Default Status**: Item tanpa status → default ke "Hutang" jika ada override parsial


````python
print("="*70)
print("CONTEXT INHERITANCE (_apply_multi_overrides)")
print("="*70)

print("\nLOGIKA:")
print("1. Sistem mencari konteks (NAMA, TANGGAL) dari entri pertama.")
print("2. Menerapkan ke entri lain yang KOSONG.")
print("3. Mendeteksi 'semua' untuk global override.")
print("4. Mendeteksi override per-item berdasarkan nama barang.")

# Simulasi langkah demi langkah
print("\n--- HASIL SETELAH CONTEXT INHERITANCE & OVERRIDE:")

temp_results = []
for i, entry in enumerate(entries, 1):
    teks_koreksi = koreksi_teks(entry)
    hasil_ekstraksi = ekstrak_entitas(teks_koreksi, teks_asli=entry)
    temp_results.append({
        "entitas": hasil_ekstraksi['entitas'],
        "original_text": entry
    })

final_results = _apply_multi_overrides(temp_results)

for i, trans in enumerate(final_results, 1):
    print(f"\n--- TRANSAKSI {i} ---")
    entitas = trans["entitas"]
    for key, value in entitas.items():
        if value not in (None, False, []):
            print(f"  {key:20s} : {value}")
````

# **E. PERHITUNGAN DATA TRANSAKSI**

Modul: `handlers/text_handler.py` (baris 1124-1204)

Setelah entitas diekstrak, sistem melakukan perhitungan:

### E1: Lookup Harga dari Master Data (baris 1125-1153)
```python
# Kode AKTUAL dari text_handler.py:
matches = cari_harga_default(ctx.db_barang, entitas["BARANG"], satuan_cari=entitas.get("SATUAN"))
if matches:
    info_harga = matches[0]
    entitas["HARGA"] = format_rupiah(info_harga["harga"])
    entitas["SATUAN"] = info_harga["satuan"]
```

### E2: Perhitungan Total (baris 1145-1153)
```python
# Kode AKTUAL:
jml_num = int(re.search(r"\d+", str(entitas["JUMLAH"])).group())
entitas["TOTAL"] = format_rupiah(jml_num * info_harga["harga"])
```

### E3: Logika Pembayaran Setengah (baris 1196-1204)
```python
# Kode AKTUAL:
if entitas.get("NOMINAL_BAYAR") == "SETENGAH" and entitas.get("TOTAL"):
    total_val = parse_rupiah(entitas["TOTAL"])
    setengah_val = total_val // 2
    entitas["NOMINAL_BAYAR"] = format_rupiah(setengah_val)
    entitas["STATUS"] = "Dicicil"
```

### E4: Fallback Default Values (baris 1210-1224)
- Jika TANGGAL kosong → default ke hari ini
- Jika ada NOMINAL_BAYAR tapi STATUS kosong → hitung: nominal < total = "Cicil", nominal ≥ total = "Lunas"


````python
print("="*70)
print("PERHITUNGAN DATA TRANSAKSI (dari text_handler.py)")
print("="*70)

# Dummy master data untuk demonstrasi perhitungan
dummy_master_barang = [
    {"nama": "Willo", "satuan": "dus", "harga": 50000},
    {"nama": "Permen Coklat", "satuan": "dus", "harga": 75000},
    {"nama": "Meses", "satuan": "bungkus", "harga": 25000}
]

print("\n--- MASTER DATA (untuk simulasi lookup) ---")
for b in dummy_master_barang:
    print(f"  {b['nama']:20s} : {format_rupiah(b['harga'])} / {b['satuan']}")

# Proses lengkap
hasil_lengkap = proses_nlp(INPUT_TEKS)

print("\n--- PERHITUNGAN PER TRANSAKSI ---")
for i, trans in enumerate(hasil_lengkap, 1):
    ent = trans["entitas"]
    barang = ent.get("BARANG")
    satuan = ent.get("SATUAN")
    
    # E1: Lookup harga (simulasi cari_harga_default)
    harga_satuan = None
    if barang and satuan:
        for b in dummy_master_barang:
            if b["nama"].lower() in str(barang).lower() and b["satuan"].lower() == str(satuan).lower():
                harga_satuan = b["harga"]
                break
    
    # E2: Hitung total
    total = None
    if harga_satuan and ent.get("JUMLAH"):
        m_jml = re.search(r"\d+", str(ent["JUMLAH"]))
        if m_jml:
            jml_num = int(m_jml.group())
            total = jml_num * harga_satuan
    
    # E3: Logika setengah
    nominal_bayar = None
    sisa_hutang = None
    if total and str(ent.get("NOMINAL_BAYAR")) == "SETENGAH":
        nominal_bayar = total // 2
        sisa_hutang = total - nominal_bayar
    elif total and ent.get("NOMINAL_BAYAR"):
        try:
            nominal_bayar = parse_rupiah(ent["NOMINAL_BAYAR"])
            sisa_hutang = total - nominal_bayar
        except:
            pass
    
    print(f"\n📊 TRANSAKSI {i} — {barang}")
    print(f"  Barang         : {barang}")
    print(f"  Jumlah         : {ent.get('JUMLAH')}")
    print(f"  Harga Satuan   : {format_rupiah(harga_satuan) if harga_satuan else '[KOSONG - lookup ke Master Data]'}")
    print(f"  TOTAL          : {format_rupiah(total) if total else '[KOSONG]'}")
    print(f"  Status         : {ent.get('STATUS')}")
    if nominal_bayar:
        print(f"  Nominal Bayar  : {format_rupiah(nominal_bayar)}")
    if sisa_hutang is not None:
        print(f"  Sisa Hutang    : {format_rupiah(sisa_hutang)}")
    print(f"  Metode         : {ent.get('METODE_PEMBAYARAN', '-')}")
````

# **F. PENYUSUNAN DATA TRANSAKSI**

Modul: `handlers/text_handler.py` (baris 1275-1286) + `services/ui_transaksi.py`

### Alur Penyusunan:

1. **Validasi Field Wajib** (baris 1277-1278):
```python
field_wajib = ["TANGGAL", "NAMA", "BARANG", "JUMLAH", "TOTAL", "STATUS", "METODE_PEMBAYARAN"]
```

2. **Jika ada field kosong** → State = `pending_confirmation`, tampilkan resume dengan tombol inline untuk mengisi data yang kurang

3. **Jika semua field lengkap** → Tampilkan resume dengan tombol `✅ Simpan` dan `❌ Batal`

4. **Simpan ke Session** (baris 1189-1194):
```python
sess = user_sessions.ensure(chat_id)
sess["hasil_nlp"] = hasil_nlp
new_entitas = {k: v for k, v in entitas.items() if v is not None and str(v).strip() != ""}
sess["entitas"] = new_entitas
```

5. **Tampilkan Resume** → `susun_balasan_resume()` di `services/ui_transaksi.py`


````python
print("="*70)
print("PENYUSUNAN DATA TRANSAKSI")
print("="*70)

field_wajib = ["TANGGAL", "NAMA", "BARANG", "JUMLAH", "TOTAL", "STATUS", "METODE_PEMBAYARAN"]

hasil_lengkap = proses_nlp(INPUT_TEKS)

for i, trans in enumerate(hasil_lengkap, 1):
    ent = trans["entitas"]
    intent = trans.get("intent", "Unknown")
    
    # Validasi field wajib
    missing = [f for f in field_wajib if not ent.get(f)]
    filled = [f for f in field_wajib if ent.get(f)]
    
    print(f"\n{'='*50}")
    print(f"TRANSAKSI {i}")
    print(f"{'='*50}")
    print(f"Intent: {intent}")
    print(f"AKSI  : {ent.get('AKSI')}")
    
    print(f"\n📋 Data Terisi ({len(filled)}/{len(field_wajib)}):")
    for f in filled:
        print(f"  ✅ {f:20s} : {ent.get(f)}")
    
    if missing:
        print(f"\n⚠️  Data Kosong ({len(missing)}):")
        for f in missing:
            print(f"  ❌ {f:20s} : [PERLU DIISI]")
        print(f"\n→ State: 'pending_confirmation'")
        print(f"→ Bot akan menampilkan resume dengan tombol inline untuk mengisi data kosong")
    else:
        print(f"\n✅ Semua field lengkap!")
        print(f"→ State: 'pending_confirmation' (expert mode)")
        print(f"→ Bot akan menampilkan resume dengan tombol ✅ Simpan dan ❌ Batal")
````

# **G. SKENARIO KHUSUS: CICILAN, LUNAS, & HUTANG**

## **Skenario 1: Pembayaran Cicilan / Setengah**

# **Kesimpulan**

Dari notebook ini, berikut rangkuman lengkap alur pemrosesan data NLP pada sistem:

## A. Preprocessing & Normalisasi (`nlp/normalizer.py`)
- **Regex** → Pembersihan karakter, normalisasi angka (0→o, 1→l), lowercase
- **Dictionary Lookup** → Pencocokan eksak pada `KAMUS_ALIAS` dan `NORMALIZATION_DICT` (typo_corrections, abbreviations)
- **RapidFuzz** (`fuzz.ratio`) → Fuzzy matching bigram (threshold ≥90) dan unigram (threshold ≥85) untuk koreksi typo

## B. Ekstraksi Entitas (`nlp/extractor.py`)
- **Regex** (`re.search`) → Ekstraksi TANGGAL, NAMA, JUMLAH, HARGA, TOTAL, SATUAN
- **Dictionary Lookup** → Deteksi BARANG (Master Data → KAMUS_ALIAS fallback)
- **Keyword Matching** → Deteksi AKSI, STATUS, METODE_PEMBAYARAN
- **RapidFuzz** (`fuzzy_intent_fallback`) → Hint intent dari dataset training

## C. Identifikasi Intent (`nlp/extractor.py` + `nlp/intent_matcher.py`)
- **RapidFuzz** (`fuzz.token_set_ratio`, threshold ≥80) → `match_intent_from_dataset()`
- **Rule-Based Fallback** → `tentukan_intent()` berdasarkan AKSI dan KONTEKS_AGREGASI

## D. Post-Processing (`nlp/processor.py`)
- **Regex** → `_apply_generic_command_guards()` untuk perintah CRUD
- **Rule-Based** → `_apply_multi_overrides()` untuk context inheritance & status override

## E. Perhitungan Data Transaksi (`handlers/text_handler.py`)
- Lookup harga dari Master Data → `cari_harga_default()`
- Perhitungan: `Total = Jumlah × Harga Satuan`
- Logika setengah: `NOMINAL_BAYAR = Total ÷ 2`
- Fallback: tanggal default hari ini, status dari perbandingan nominal vs total

## F. Penyusunan Data Transaksi (`handlers/text_handler.py`)
- Validasi field wajib: `[TANGGAL, NAMA, BARANG, JUMLAH, TOTAL, STATUS, METODE_PEMBAYARAN]`
- Simpan ke session → Tampilkan resume via `susun_balasan_resume()`
- Jika ada field kosong → tombol inline untuk mengisi
- Jika lengkap → tombol ✅ Simpan dan ❌ Batal
