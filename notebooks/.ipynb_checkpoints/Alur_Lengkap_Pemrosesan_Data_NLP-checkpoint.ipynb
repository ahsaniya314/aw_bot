# **A. PREPROCESSING & NORMALISASI**

---

## **Langkah A1: Input Data**

Menerima teks mentah dari pengguna.

```python
# Contoh input dengan typo dan multiple entri
INPUT_TEKS = "pak anDi Pesan prmen1 10 dus, mses 5 bks, will0 3 dus belum lunwas semua"

print("="*70)
print("INPUT DATA MENTAH")
print("="*70)
print(f"\nTeks: {INPUT_TEKS}")
```

---

## **Langkah A2: Pemisahan Multi-Entry**

Memisahkan teks yang berisi lebih dari satu transaksi menjadi beberapa entri terpisah.

```python
entries = split_multi_entries(INPUT_TEKS)

print("="*70)
print("HASIL PEMISAHAN MULTI-ENTRY")
print("="*70)
print(f"\nJumlah entri: {len(entries)}")

for i, entry in enumerate(entries, 1):
    print(f"\nEntri {i}:")
    print(f"  Teks: {entry}")
```

---

## **Langkah A3: Text Cleaning**

Membersihkan teks dari karakter yang tidak dibutuhkan (seperti tanda baca berlebih, karakter khusus, dsb).

```python
# Kita ambil entri pertama untuk demonstrasi detail
entri_contoh = entries[0]

print("="*70)
print("TEXT CLEANING")
print("="*70)

print(f"\n1. Teks asli:        {entri_contoh}")

# Contoh text cleaning sederhana: menghapus karakter non-alfanumerik kecuali spasi
teks_clean = re.sub(r'[^\w\s]', '', entri_contoh)
print(f"2. Setelah cleaning:  {teks_clean}")
```

---

## **Langkah A4: Lowercase**

Mengubah semua huruf menjadi huruf kecil agar perhitungan fuzzy matching menjadi konsisten.

```python
print("="*70)
print("LOWERCASE")
print("="*70)

print(f"\n1. Teks sebelumnya: {teks_clean}")

# Ubah semua huruf menjadi lowercase
teks_lower = teks_clean.lower()
teks_bersih = teks_lower

print(f"2. Setelah lowercase: {teks_lower}")
```

---

## **Langkah A5: Tokenization**

Memisahkan teks menjadi token (kata-kata) individual.

```python
tokens = teks_bersih.split()

print("="*70)
print("TOKENIZATION")
print("="*70)

print(f"\nTeks: {teks_bersih}")
print(f"\nTokens:")
for i, token in enumerate(tokens, 1):
    print(f"  Token {i}: '{token}'")
```

---

## **Langkah A6: Normalization (Kamus Slang & Singkatan)**

Mengubah kata slang/singkatan menjadi kata baku.

```python
print("="*70)
print("KAMUS NORMALISASI")
print("="*70)

print("\n📦 Contoh KAMUS_ALIAS (nama barang):")
contoh_barang = list(KAMUS_ALIAS.items())[:8]
for typo, benar in contoh_barang:
    print(f"  '{typo}' → '{benar}'")
```

---

# **B. PROSESING DATA: FUZZY MATCHING → TRANSAKSI**

---

## **Langkah B1 & B2: Fuzzy Matching (Levenshtein Distance & Similarity Score)**

Koreksi typo dengan menghitung jarak edit dan skor kesamaan.

```python
print("="*70)
print("LEVENSHTEIN DISTANCE & SIMILARITY SCORE")
print("="*70)

# Demonstrasi dengan rapidfuzz
from rapidfuzz import fuzz

contoh_typo = [
    ("prmen", "permen"),
    ("mses", "meses"),
    ("willo", "willo")
]

for typo, benar in contoh_typo:
    similarity = fuzz.ratio(typo, benar)
    print(f"\n• '{typo}' vs '{benar}':")
    print(f"  Similarity: {similarity}%")
    print(f"  Status:     {'✅ Mirip' if similarity >= 85 else '❌ Tidak mirip'}")
```

```python
# Koreksi teks lengkap
print("\n" + "="*70)
print("HASIL KOREKSI TEKS LENGKAP")
print("="*70)

for i, entry in enumerate(entries, 1):
    teks_dikoreksi = koreksi_teks(entry)
    
    print(f"\nEntri {i}:")
    print(f"  Sebelum: '{entry}'")
    print(f"  Sesudah: '{teks_dikoreksi}'")
```

---

## **Langkah B3: Entity Extraction dengan Regex**

Mengekstrak entitas penting seperti tanggal, nama, barang, jumlah, harga, dll.

```python
print("="*70)
print("ENTITY EXTRACTION DENGAN REGEX")
print("="*70)

for i, entry in enumerate(entries, 1):
    teks_koreksi = koreksi_teks(entry)
    hasil_ekstraksi = ekstrak_entitas(teks_koreksi, teks_asli=entry)
    
    print(f"\n--- ENTITAS ENTRI {i} ---")
    print(f"Teks asli   : {entry}")
    print(f"Teks koreksi: {teks_koreksi}")
    
    entitas = hasil_ekstraksi['entitas']
    for key, value in entitas.items():
        if value not in (None, False, []):
            print(f"  {key:20s} : {value}")
```

---

## **Langkah B4: Context Inheritance (Warisan Konteks) - Penting untuk Laporan!**

Menerapkan konteks (nama, tanggal) dari entri pertama ke entri lain yang kosong.

```python
print("="*70)
print("CONTEXT INHERITANCE (WARISAN KONTEKS)")
print("="*70)

print("\nLOGIKA:")
print("1. Sistem mencari konteks (nama, tanggal) dari entri pertama yang memilikinya.")
print("2. Lalu menerapkan ke entri lain yang KOSONG.")
print("3. Ini untuk kasus: SATU PESANAN, SATU NAMA PEMBELI!")

# Simulasi langkah demi langkah
print("\n--- HASIL SETELAH CONTEXT INHERITANCE:")

# 1. Buat list hasil sementara
temp_results = []
for i, entry in enumerate(entries, 1):
    teks_koreksi = koreksi_teks(entry)
    hasil_ekstraksi = ekstrak_entitas(teks_koreksi, teks_asli=entry)
    temp_results.append({
        "entitas": hasil_ekstraksi['entitas'],
        "original_text": entry
    })

# 2. Terapkan _apply_multi_overrides
final_results = _apply_multi_overrides(temp_results)

for i, trans in enumerate(final_results, 1):
    print(f"\n--- TRANSAKSI {i} ---")
    entitas = trans["entitas"]
    for key, value in entitas.items():
        if value not in (None, False, []):
            print(f"  {key:20s} : {value}")
```

---

## **Langkah B5: Perhitungan Total Harga & Penanganan Data Kosong**

Menghitung total harga dan menangani data yang hilang.

```python
print("="*70)
print("PERHITUNGAN TOTAL HARGA & PENANGANAN DATA KOSONG")
print("="*70)

print("\n--- LOGIKA PERHITUNGAN TOTAL ---")
print("1. Cari harga satuan dari Master Data.")
print("2. Hitung: Total = Jumlah × Harga Satuan.")
print("3. Jika barang tidak ditemukan, harga kosong.")
print("4. Jika status 'setengah' → Total ÷ 2.")

# Dummy master data untuk demonstrasi
dummy_master_barang = [
    {"nama": "Willo", "satuan": "dus", "harga": 50000},
    {"nama": "Permen", "satuan": "dus", "harga": 75000},
    {"nama": "Meses", "satuan": "bks", "harga": 25000}
]

# Simulasi perhitungan
temp_results = []
for i, entry in enumerate(entries, 1):
    teks_koreksi = koreksi_teks(entry)
    hasil_ekstraksi = ekstrak_entitas(teks_koreksi, teks_asli=entry)
    temp_results.append({
        "entitas": hasil_ekstraksi['entitas'],
        "original_text": entry
    })

final_results = _apply_multi_overrides(temp_results)

print("\n--- HASIL PERHITUNGAN TOTAL:")
for i, trans in enumerate(final_results, 1):
    ent = trans["entitas"]
    
    # Cari harga
    barang = ent.get("BARANG")
    satuan = ent.get("SATUAN")
    harga_satuan = None
    
    if barang and satuan:
        for b in dummy_master_barang:
            if b["nama"].lower() in barang.lower() and b["satuan"].lower() == satuan.lower():
                harga_satuan = b["harga"]
                break
    
    # Hitung total
    total = None
    if harga_satuan and ent.get("JUMLAH"):
        m_jml = re.search(r"\d+", str(ent["JUMLAH"]))
        if m_jml:
            jml_num = int(m_jml.group())
            total = jml_num * harga_satuan
    
    # Hitung nominal bayar jika status 'setengah'
    nominal_bayar = None
    if total and str(ent.get("NOMINAL_BAYAR")) == "SETENGAH":
        nominal_bayar = total // 2
    
    # Tampilkan
    print(f"\nTRANSAKSI {i} - {barang}")
    print(f"  Barang        : {barang}")
    print(f"  Jumlah        : {ent.get('JUMLAH')}")
    print(f"  Harga Satuan  : {format_rupiah(harga_satuan) if harga_satuan else '[KOSONG - cari harga!]'}")
    print(f"  TOTAL         : {format_rupiah(total) if total else '[KOSONG]'}")
    print(f"  Status        : {ent.get('STATUS')}")
    if nominal_bayar:
        print(f"  Nominal Bayar : {format_rupiah(nominal_bayar)}")
        print(f"  Sisa Hutang   : {format_rupiah(total - nominal_bayar)}")
```

# **C. SKENARIO KHUSUS: CICILAN, LUNAS, & HUTANG**

---

## **Skenario 1: Pembayaran Cicilan / Setengah**

```python
print("="*70)
print("SKENARIO: PEMBAYARAN CICILAN / SETENGAH")
print("="*70)

input_cicilan = "pak andi pesan willo 4 dus setengah"
hasil_cicilan = proses_nlp(input_cicilan)

if hasil_cicilan:
    ent = hasil_cicilan[0]['entitas']
    # Kita simulasi cari harga
    harga_satuan = 50000
    total = 4 * harga_satuan
    nominal_bayar = total // 2
    sisa = total - nominal_bayar
    
    print(f"\nInput: {input_cicilan}")
    print(f"\n📊 Detail Transaksi:")
    print(f"  Barang        : {ent.get('BARANG')}")
    print(f"  Jumlah        : {ent.get('JUMLAH')}")
    print(f"  Harga Satuan  : {format_rupiah(harga_satuan)}")
    print(f"  TOTAL         : {format_rupiah(total)}")
    print(f"\n💰 Pembayaran:")
    print(f"  Nominal Bayar : {format_rupiah(nominal_bayar)}")
    print(f"  Sisa Hutang   : {format_rupiah(sisa)}")
    print(f"  Status        : {ent.get('STATUS')}")
```

---

## **Skenario 2: Pembayaran Lunas**

```python
print("="*70)
print("SKENARIO: PEMBAYARAN LUNAS")
print("="*70)

input_lunas = "pak andi pesan willo 4 dus lunas tunai"
hasil_lunas = proses_nlp(input_lunas)

if hasil_lunas:
    ent = hasil_lunas[0]['entitas']
    harga_satuan = 50000
    total = 4 * harga_satuan
    
    print(f"\nInput: {input_lunas}")
    print(f"\n📊 Detail Transaksi:")
    print(f"  Barang        : {ent.get('BARANG')}")
    print(f"  Jumlah        : {ent.get('JUMLAH')}")
    print(f"  TOTAL         : {format_rupiah(total)}")
    print(f"\n💰 Pembayaran:")
    print(f"  Status        : {ent.get('STATUS')}")
    print(f"  Metode        : {ent.get('METODE_PEMBAYARAN')}")
    print(f"  Sisa Hutang   : Rp 0")
```
