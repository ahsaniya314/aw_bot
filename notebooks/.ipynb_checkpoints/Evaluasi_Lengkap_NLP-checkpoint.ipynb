# Evaluasi Lengkap Sistem NLP & Intent Identification

Notebook ini melakukan evaluasi:
1. **Identifikasi Intent** - Confusion Matrix, Accuracy, Precision, Recall, F1-Score
2. **Ekstraksi Entitas** - Precision, Recall, F1-Score per entity type

## 1. Setup Environment & Import Library

```python
import sys
from pathlib import Path
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Tambahkan root direktori proyek ke sys.path
ROOT = Path().resolve().parent
sys.path.insert(0, str(ROOT))

print(f"Root proyek: {ROOT}")
```

## 2. Muat Data Latih & Fungsi NLP

```python
from nlp.intent_matcher import match_intent_from_dataset, DATASET_TO_SYSTEM_INTENT
from nlp.embedded_data import NLP_TRAINING_EXAMPLES
from nlp.processor import proses_nlp

print(f"Total training examples: {len(NLP_TRAINING_EXAMPLES)}")
```

## 3. Evaluasi Identifikasi Intent

```python
# Siapkan data evaluasi
y_true = []
y_pred = []
prediction_details = []

for example in NLP_TRAINING_EXAMPLES:
    text = example.get("text", "")
    actual_tag = example.get("intent", "")

    if not text or not actual_tag:
        continue

    actual_intent = DATASET_TO_SYSTEM_INTENT.get(actual_tag, actual_tag)
    predicted_intent, score = match_intent_from_dataset(text)

    y_true.append(actual_intent)
    y_pred.append(predicted_intent if predicted_intent else "Unknown")

    prediction_details.append({
        "text": text,
        "actual": actual_intent,
        "predicted": predicted_intent or "Unknown",
        "score": score,
        "is_correct": predicted_intent == actual_intent
    })

print(f"Total examples evaluated: {len(y_true)}")
```

```python
# Hitung akurasi
correct_count = sum(1 for p in prediction_details if p["is_correct"])
accuracy = (correct_count / len(y_true)) * 100 if len(y_true) > 0 else 0

print("=" * 80)
print("HASIL EVALUASI INTENT IDENTIFICATION")
print("=" * 80)
print(f"Total examples: {len(y_true)}")
print(f"Correct predictions: {correct_count}")
print(f"Incorrect predictions: {len(y_true) - correct_count}")
print(f"Accuracy: {accuracy:.2f}%")
print("=" * 80)
```

```python
# Tampilkan classification report
print("Classification Report:")
print("=" * 80)
try:
    report = classification_report(y_true, y_pred, output_dict=True)
    print(classification_report(y_true, y_pred))
    
    # Plot pie chart akurasi
    plt.figure(figsize=(8, 5))
    labels = ['Benar', 'Salah']
    sizes = [correct_count, len(y_true) - correct_count]
    colors = ['#4CAF50', '#F44336']
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title('Akurasi Model Intent Identification')
    plt.axis('equal')
    plt.show()
    
    # Plot confusion matrix
    # Dapatkan semua unique labels
    all_labels = sorted(list(set(y_true + y_pred)))
    cm = confusion_matrix(y_true, y_pred, labels=all_labels)
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=all_labels, yticklabels=all_labels)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix - Intent Identification')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"Warning: Could not generate full report: {e}")
```

```python
# Tampilkan contoh prediksi salah
incorrect = [p for p in prediction_details if not p["is_correct"]]

if incorrect:
    print("\n" + "=" * 80)
    print(f"CONTOH PREDIKSI SALAH ({len(incorrect)} contoh):")
    print("=" * 80)
    for i, p in enumerate(incorrect[:10]):  # Tampilkan 10 contoh pertama
        print(f"\n{i+1}. Teks: {p['text']}")
        print(f"   Actual Intent: {p['actual']}")
        print(f"   Predicted Intent: {p['predicted']}")
        print(f"   Score: {p['score']:.2f}")
```

## 4. Evaluasi Ekstraksi Entitas

```python
# Daftar entity types yang ingin kita evaluasi
ENTITY_TYPES = [
    "TANGGAL", 
    "NAMA", 
    "BARANG", 
    "JUMLAH", 
    "SATUAN", 
    "STATUS", 
    "METODE_PEMBAYARAN"
]

# Fungsi untuk mengevaluasi ekstraksi entitas
def evaluate_entity_extraction():
    entity_metrics = defaultdict(lambda: {'TP': 0, 'FP': 0, 'FN': 0})
    extraction_details = []
    
    # Kita gunakan NLP_TRAINING_EXAMPLES untuk evaluasi
    # Catatan: Kita asumsikan NLP_TRAINING_EXAMPLES memiliki field 'entities' (jika tidak ada, kita skip)
    for example in NLP_TRAINING_EXAMPLES:
        text = example.get("text", "")
        
        if not text:
            continue
        
        # Dapatkan ground truth entities dari example (jika ada)
        actual_entities = example.get("entities", {})
        
        # Dapatkan predicted entities dari sistem
        try:
            results = proses_nlp(text)
            if results:
                predicted_entities = results[0].get("entitas", {})
            else:
                predicted_entities = {}
        except Exception as e:
            predicted_entities = {}
            print(f"Warning: Error processing '{text}': {e}")
        
        extraction_details.append({
            "text": text,
            "actual": actual_entities,
            "predicted": predicted_entities
        })
        
        # Hitung TP, FP, FN untuk setiap entity type
        for entity_type in ENTITY_TYPES:
            actual_val = str(actual_entities.get(entity_type, "")).strip().lower() if actual_entities.get(entity_type) else ""
            predicted_val = str(predicted_entities.get(entity_type, "")).strip().lower() if predicted_entities.get(entity_type) else ""
            
            if actual_val and predicted_val and actual_val == predicted_val:
                entity_metrics[entity_type]['TP'] += 1
            elif actual_val and (not predicted_val or predicted_val != actual_val):
                entity_metrics[entity_type]['FN'] += 1
            elif predicted_val and not actual_val:
                entity_metrics[entity_type]['FP'] += 1
    
    return entity_metrics, extraction_details

# Jalankan evaluasi
entity_metrics, extraction_details = evaluate_entity_extraction()
print(f"Evaluasi selesai untuk {len(extraction_details)} examples")
```

```python
# Hitung dan tampilkan metrik per entity
print("=" * 80)
print("HASIL EVALUASI EKSTRAKSI ENTITAS")
print("=" * 80)

entity_results = []

for entity_type in ENTITY_TYPES:
    TP = entity_metrics[entity_type]['TP']
    FP = entity_metrics[entity_type]['FP']
    FN = entity_metrics[entity_type]['FN']
    
    # Hitung Precision, Recall, F1-Score
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0.0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0
    
    entity_results.append({
        "Entity Type": entity_type,
        "TP": TP,
        "FP": FP,
        "FN": FN,
        "Precision": f"{precision:.2f}",
        "Recall": f"{recall:.2f}",
        "F1-Score": f"{f1_score:.2f}"
    })
    
    print(f"\n{entity_type}:")
    print(f"  TP (True Positives): {TP}")
    print(f"  FP (False Positives): {FP}")
    print(f"  FN (False Negatives): {FN}")
    print(f"  Precision: {precision:.2f}")
    print(f"  Recall: {recall:.2f}")
    print(f"  F1-Score: {f1_score:.2f}")

# Tampilkan sebagai tabel
entity_df = pd.DataFrame(entity_results)
print("\n" + "=" * 80)
print("RINGKASAN METRIK EKSTRAKSI ENTITAS:")
print("=" * 80)
print(entity_df.to_string(index=False))
```

```python
# Plot metrik per entity
plot_data = []
for entity in entity_results:
    plot_data.append({
        "Entity Type": entity["Entity Type"],
        "Metric": "Precision",
        "Value": float(entity["Precision"])
    })
    plot_data.append({
        "Entity Type": entity["Entity Type"],
        "Metric": "Recall",
        "Value": float(entity["Recall"])
    })
    plot_data.append({
        "Entity Type": entity["Entity Type"],
        "Metric": "F1-Score",
        "Value": float(entity["F1-Score"])
    })

plot_df = pd.DataFrame(plot_data)

plt.figure(figsize=(14, 8))
sns.barplot(x="Entity Type", y="Value", hue="Metric", data=plot_df, palette="viridis")
plt.title('Perbandingan Metrik Ekstraksi Entitas')
plt.xlabel('Tipe Entitas')
plt.ylabel('Nilai (0-1)')
plt.ylim(0, 1.05)
plt.xticks(rotation=45)
plt.legend(title="Metrik")
plt.tight_layout()
plt.show()
```

```python
# Tampilkan contoh detail ekstraksi entitas
print("\n" + "=" * 80)
print("CONTOH DETAIL EKSTRAKSI ENTITAS (5 contoh pertama):")
print("=" * 80)
for i, detail in enumerate(extraction_details[:5]):
    print(f"\n{i+1}. Teks: {detail['text']}")
    print(f"   Actual Entities: {detail['actual']}")
    print(f"   Predicted Entities: {detail['predicted']}")
```

## 5. Simpan Hasil Evaluasi (Opsional)

```python
# Simpan ke file CSV untuk analisis lebih lanjut
from datetime import datetime

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# Simpan prediction details intent
intent_df = pd.DataFrame(prediction_details)
intent_file = ROOT / f"logs/intent_evaluation_{timestamp}.csv"
intent_df.to_csv(intent_file, index=False)
print(f"Intent evaluation saved to: {intent_file}")

# Simpan entity evaluation
entity_file = ROOT / f"logs/entity_evaluation_{timestamp}.csv"
entity_df.to_csv(entity_file, index=False)
print(f"Entity evaluation saved to: {entity_file}")
```
