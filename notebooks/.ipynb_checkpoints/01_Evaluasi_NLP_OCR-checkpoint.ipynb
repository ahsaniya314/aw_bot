## 2. Evaluasi Model NLP

Kita evaluasi performa fuzzy matching untuk intent classification.

```python
from nlp.intent_matcher import match_intent_from_dataset, DATASET_TO_SYSTEM_INTENT
from nlp.embedded_data import NLP_TRAINING_EXAMPLES

print(f"Total data NLP: {len(NLP_TRAINING_EXAMPLES)}")
```

```python
# Evaluasi NLP
total = 0
correct = 0
wrong = []
confusion = defaultdict(lambda: defaultdict(int))

for example in NLP_TRAINING_EXAMPLES:
    text = example.get("text", "")
    actual_tag = example.get("intent", "")

    if not text or not actual_tag:
        continue

    actual_intent = DATASET_TO_SYSTEM_INTENT.get(actual_tag, actual_tag)

    total += 1
    predicted_intent, score = match_intent_from_dataset(text)

    confusion[actual_intent][predicted_intent or "Unknown"] += 1

    if predicted_intent == actual_intent:
        correct += 1
    else:
        wrong.append({
            "text": text,
            "actual": actual_intent,
            "predicted": predicted_intent,
            "score": score
        })

accuracy = (correct / total) * 100 if total > 0 else 0

print(f"Total data uji     : {total}")
print(f"Prediksi benar     : {correct}")
print(f"Prediksi salah     : {len(wrong)}")
print(f"Akurasi            : {accuracy:.2f}%")
```

```python
# Plot akurasi
plt.figure(figsize=(8, 5))
labels = ['Benar', 'Salah']
sizes = [correct, len(wrong)]
colors = ['#4CAF50', '#F44336']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Akurasi Model NLP')
plt.axis('equal')
plt.show()
```

```python
# Tampilkan 10 contoh salah prediksi
print("Contoh salah prediksi:")
for i, w in enumerate(wrong[:10], 1):
    print(f"\n{i}. {w['text']}")
    print(f"   Actual: {w['actual']}")
    print(f"   Predicted: {w['predicted']}, Score: {w['score']}")
```

```python
# Buat confusion matrix
intents = sorted(list(confusion.keys()))
confusion_df = pd.DataFrame(0, index=intents, columns=intents + ['Unknown'])

for actual, preds in confusion.items():
    for pred, count in preds.items():
        if pred in confusion_df.columns:
            confusion_df.loc[actual, pred] = count

print("Confusion Matrix:")
confusion_df
```
