# Analisis Dataset Lengkap untuk Laporan Penelitian

Notebook ini berisi analisis menyeluruh tentang semua dataset yang digunakan dalam sistem bot!

## Daftar Isi:
1. **Dataset NLP Normalization**: Kamus koreksi typo, sinonim, dan singkatan
2. **Dataset Intent Patterns**: Contoh kalimat untuk setiap intent
3. **Dataset Training NLP**: Contoh kalimat beserta entity yang diekstrak
4. **Valid Units**: Satuan produk yang dikenal sistem
5. **Analisis Statistik dan Visualisasi**

---

## 1. Setup dan Import Library

```python
import sys
import json
import csv
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Set style untuk plot
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Tambahkan root proyek ke path
ROOT = Path().resolve().parent
sys.path.insert(0, str(ROOT))
FILES_DIR = ROOT / "files"

print(f"Folder dataset: {FILES_DIR}")
print(f"File yang tersedia: {[f.name for f in FILES_DIR.glob('*')]}")
```

---

## 2. Analisis Dataset NLP Normalization (`nlp_normalization_dict.json`)

```python
# Load dataset
with open(FILES_DIR / "nlp_normalization_dict.json", "r", encoding="utf-8") as f:
    norm_dict = json.load(f)

print("=== Struktur Dataset Normalization ===")
print(f"- Metadata: {norm_dict.get('metadata', {})}")
print(f"- Jumlah typo corrections: {len(norm_dict.get('typo_corrections', {}))}")
print(f"- Jumlah sinonim groups: {len(norm_dict.get('synonyms', {}))}")
print(f"- Jumlah singkatan: {len(norm_dict.get('abbreviations', {}))}")
print(f"- Jumlah entity patterns: {len(norm_dict.get('entity_patterns', {}))}")
```

```python
# Lihat 5 contoh typo corrections
print("\n=== Contoh 5 Typo Corrections ===")
typo_items = list(norm_dict.get('typo_corrections', {}).items())
for typo, correct in typo_items[:5]:
    print(f"  '{typo}' -> '{correct}'")
```

```python
# Hitung dan visualisasikan jumlah kata per sinonim group
synonyms = norm_dict.get('synonyms', {})
synonym_counts = {k: len(v) for k, v in synonyms.items()}

print("\n=== Jumlah Kata per Sinonim Group ===")
synonym_df = pd.DataFrame(list(synonym_counts.items()), columns=["Group", "Jumlah Kata"])
display(synonym_df.sort_values("Jumlah Kata", ascending=False))

# Plot
plt.figure(figsize=(14, 7))
sns.barplot(data=synonym_df, x="Group", y="Jumlah Kata", palette="viridis")
plt.title("Jumlah Kata per Sinonim Group")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

---

## 3. Analisis Dataset Intent Patterns (`nlp_dataset.json`)

```python
# Load dataset
with open(FILES_DIR / "nlp_dataset.json", "r", encoding="utf-8") as f:
    intent_data = json.load(f)

intents = intent_data.get("intents", [])
print(f"Total jumlah intent: {len(intents)}")
```

```python
# Hitung jumlah contoh per intent
intent_stats = []
for intent in intents:
    tag = intent.get("tag", "unknown")
    patterns = intent.get("patterns", [])
    intent_stats.append({"Intent": tag, "Jumlah Contoh": len(patterns)})

intent_df = pd.DataFrame(intent_stats)
print("=== Statistik Intent ===")
display(intent_df.sort_values("Jumlah Contoh", ascending=False))

# Plot distribusi jumlah contoh per intent
plt.figure(figsize=(16, 8))
sns.barplot(data=intent_df, x="Intent", y="Jumlah Contoh", palette="coolwarm")
plt.title("Distribusi Jumlah Contoh Kalimat per Intent")
plt.xticks(rotation=60, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Lihat 3 contoh intent dengan contoh kalimatnya
print("\n=== Contoh Intent dan Kalimat ===")
for intent in intents[:3]:
    print(f"\nIntent: {intent['tag']}")
    print(f"Contoh kalimat: {intent['patterns'][:3]}")
```

---

## 4. Analisis Dataset Training NLP (`nlp_dataset.csv`)

```python
# Load dataset CSV
training_df = pd.read_csv(FILES_DIR / "nlp_dataset.csv")
print(f"Total baris data: {len(training_df)}")
display(training_df.head(10))
```

```python
# Distribusi intent di training dataset
plt.figure(figsize=(16, 8))
sns.countplot(data=training_df, y="intent", order=training_df["intent"].value_counts().index, palette="pastel")
plt.title("Distribusi Intent di Dataset Training")
plt.xlabel("Jumlah Contoh")
plt.tight_layout()
plt.show()
```

```python
# Analisis panjang kalimat
training_df["Panjang Kalimat"] = training_df["text"].apply(lambda x: len(str(x).split()))

print("=== Statistik Panjang Kalimat ===")
display(training_df["Panjang Kalimat"].describe())

plt.figure(figsize=(12, 6))
sns.histplot(data=training_df, x="Panjang Kalimat", bins=20, kde=True, color="skyblue")
plt.title("Distribusi Panjang Kalimat (Jumlah Kata)")
plt.xlabel("Jumlah Kata")
plt.tight_layout()
plt.show()
```

---

## 5. Analisis Valid Units (`valid_units.txt`)

```python
# Load valid units
with open(FILES_DIR / "valid_units.txt", "r", encoding="utf-8") as f:
    valid_units = [line.strip() for line in f.readlines() if line.strip()]

print(f"Total satuan valid: {len(valid_units)}")
print(f"\nDaftar satuan: {valid_units}")
```

---

## 6. Analisis Gabungan dan Kesimpulan

```python
# Hitung total statistik
total_intents = len(intents)
total_patterns = sum(len(i["patterns"]) for i in intents)
total_typos = len(typo_items)
total_synonym_groups = len(synonyms)
total_training_examples = len(training_df)
total_units = len(valid_units)

summary_data = {
    "Kategori": [
        "Jumlah Intent",
        "Total Contoh Kalimat Intent",
        "Jumlah Typo Corrections",
        "Jumlah Sinonim Groups",
        "Jumlah Contoh Training (Entity Labeled)",
        "Jumlah Satuan Valid"
    ],
    "Jumlah": [
        total_intents,
        total_patterns,
        total_typos,
        total_synonym_groups,
        total_training_examples,
        total_units
    ]
}

summary_df = pd.DataFrame(summary_data)
print("=== RINGKASAN STATISTIK DATASET ===")
display(summary_df)
```

```python
# Visualisasi ringkasan
plt.figure(figsize=(14, 8))
colors = sns.color_palette("pastel", len(summary_df))
bars = plt.bar(summary_df["Kategori"], summary_df["Jumlah"], color=colors)
plt.title("Ringkasan Statistik Seluruh Dataset")
plt.ylabel("Jumlah")
plt.xticks(rotation=45, ha='right')

# Tambahkan label angka di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
