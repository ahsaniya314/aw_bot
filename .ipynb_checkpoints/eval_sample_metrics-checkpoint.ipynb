# Evaluasi Intent & Entitas (Accuracy, Precision, Recall, F1)

Notebook ini menjalankan pipeline NLP project BOT untuk input contoh, lalu menghitung metrik evaluasi:
- Intent: accuracy
- Entitas (slot-based): slot_accuracy, precision, recall, f1


```python
import json
from typing import Any, Dict, Iterable, List, Optional, Sequence, Tuple

from nlp.processor import proses_nlp


def _norm(v: Any) -> Optional[str]:
    if v is None:
        return None
    if isinstance(v, bool):
        return str(v).lower()
    s = str(v)
    s = " ".join(s.strip().split())
    return s.lower() if s else None


def _pairs(entitas: Dict[str, Any], fields: Iterable[str]) -> List[Tuple[str, str]]:
    out: List[Tuple[str, str]] = []
    for f in fields:
        if f not in entitas:
            continue
        nv = _norm(entitas.get(f))
        if nv is None:
            continue
        out.append((f, nv))
    return out


def _safe_div(a: float, b: float) -> float:
    return a / b if b else 0.0


def _f1(p: float, r: float) -> float:
    return _safe_div(2 * p * r, p + r)


def evaluate_intent(y_true: Sequence[str], y_pred: Sequence[str]) -> Dict[str, float]:
    n = min(len(y_true), len(y_pred))
    correct = sum(1 for i in range(n) if (y_true[i] or "").strip() == (y_pred[i] or "").strip())
    acc = _safe_div(correct, n)
    return {"accuracy": acc, "n": float(n), "correct": float(correct)}


def evaluate_entities_slot_f1(
    true_items: Sequence[Dict[str, Any]],
    pred_items: Sequence[Dict[str, Any]],
    fields: Sequence[str],
) -> Dict[str, float]:
    n = min(len(true_items), len(pred_items))

    tp = 0
    fp = 0
    fn = 0
    expected_slots = 0
    correct_slots = 0

    for i in range(n):
        true_pairs = set(_pairs(true_items[i], fields))
        pred_pairs = set(_pairs(pred_items[i], fields))

        expected_slots += len(true_pairs)
        inter = true_pairs & pred_pairs
        correct_slots += len(inter)

        tp += len(inter)
        fp += len(pred_pairs - true_pairs)
        fn += len(true_pairs - pred_pairs)

    precision = _safe_div(tp, tp + fp)
    recall = _safe_div(tp, tp + fn)
    f1 = _f1(precision, recall)
    slot_accuracy = _safe_div(correct_slots, expected_slots)

    return {
        "slot_accuracy": slot_accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "tp": float(tp),
        "fp": float(fp),
        "fn": float(fn),
        "items": float(n),
        "expected_slots": float(expected_slots),
        "correct_slots": float(correct_slots),
    }


def details_per_item(
    true_items: Sequence[Dict[str, Any]],
    pred_items: Sequence[Dict[str, Any]],
    fields: Sequence[str],
) -> List[Dict[str, Any]]:
    n = min(len(true_items), len(pred_items))
    rows: List[Dict[str, Any]] = []
    for i in range(n):
        row: Dict[str, Any] = {"item": i + 1}
        for f in fields:
            row[f"true_{f}"] = true_items[i].get(f)
            row[f"pred_{f}"] = pred_items[i].get(f)
            row[f"match_{f}"] = _norm(true_items[i].get(f)) == _norm(pred_items[i].get(f))
        rows.append(row)
    return rows
```

```python
text_input = "\n".join(
    [
        "19-06-2026 Nama : Budi budis",
        "bembeng 40 ctn",
        "meses 20 ctn",
        "twilo 100 ctn",
    ]
)

pred = proses_nlp(text_input, already_normalized=True)
print(json.dumps(pred, indent=2, ensure_ascii=False))
```

```python
ground_truth = [
    {
        "intent": "Catat_Penjualan_Lunas",
        "entitas": {
            "TANGGAL": "19-06-2026",
            "NAMA": "Budi Budis",
            "BARANG": "Bembeng",
            "JUMLAH": "40 karton",
            "SATUAN": "karton",
        },
    },
    {
        "intent": "Catat_Penjualan_Lunas",
        "entitas": {
            "TANGGAL": "19-06-2026",
            "NAMA": "Budi Budis",
            "BARANG": "Meses",
            "JUMLAH": "20 karton",
            "SATUAN": "karton",
        },
    },
    {
        "intent": "Catat_Penjualan_Lunas",
        "entitas": {
            "TANGGAL": "19-06-2026",
            "NAMA": "Budi Budis",
            "BARANG": "Willo",
            "JUMLAH": "100 karton",
            "SATUAN": "karton",
        },
    },
]

fields = ["TANGGAL", "NAMA", "BARANG", "JUMLAH", "SATUAN"]

y_true_intent = [x.get("intent") for x in ground_truth]
y_pred_intent = [x.get("intent") for x in pred]

intent_metrics = evaluate_intent(y_true_intent, y_pred_intent)

true_entities = [x.get("entitas") or {} for x in ground_truth]
pred_entities = [x.get("entitas") or {} for x in pred]
entity_metrics = evaluate_entities_slot_f1(true_entities, pred_entities, fields)

result = {
    "metrics": {
        "intent": intent_metrics,
        "entities_slot": entity_metrics,
    },
    "details_per_item": details_per_item(true_entities, pred_entities, fields),
}

print(json.dumps(result, indent=2, ensure_ascii=False))
```
