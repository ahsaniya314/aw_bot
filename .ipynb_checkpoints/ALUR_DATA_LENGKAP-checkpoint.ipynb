# 📊 Alur Lengkap Pemrosesan Data NLP (VERSI LENGKAP)

Notebook ini menunjukkan SEMUA langkah secara terperinci sesuai permintaan kamu:
1. Input Data
2. Text Cleaning
3. Lowercase
4. Tokenization
5. Normalization
6. Fuzzy Matching (Dictionary Lookup, Levenshtein Distance, Similarity Score)
7. Ekstraksi Entitas dengan Regex
8. Identifikasi Intent
9. Penyusunan Data Transaksi

---

```python
# ✅ Langkah 0: Import semua modul
import sys
import os
import re
import pprint
sys.path.insert(0, os.path.abspath('.'))

from nlp.processor import proses_nlp
from nlp.normalizer import koreksi_teks
from nlp.extractor import ekstrak_entitas
from nlp.dictionaries import KAMUS_ALIAS, NORMALIZATION_DICT

print("✅ Semua modul berhasil diimport!")
```

---

```python
# 📥 Langkah a: Input Data
input_text = "hari ini pak andi ambil willo 10 karto lunas dibayar tf"

print("="*60)
print("📥 LANGKAH a: INPUT DATA (DATA MENTAH DARI USER)")
print("="*60)
print(f"Input: {input_text}")
```

---

```python
# 🧹 Langkah b: Text Cleaning
text_clean = re.sub(r'[^\w\s]', '', input_text)
text_clean = re.sub(r'\s+', ' ', text_clean).strip()

print("="*60)
print("🧹 LANGKAH b: TEXT CLEANING (MEMBERSIHKAN TEKS)")
print("="*60)
print(f"- Sebelum: {input_text}")
print(f"- Sesudah: {text_clean}")
print("\nPenjelasan: Menghapus karakter khusus dan spasi berlebih")
```

---

```python
# 🔤 Langkah c: Lowercase
text_lower = text_clean.lower()

print("="*60)
print("🔤 LANGKAH c: LOWERCASE (MENGUBAH KE HURUF KECIL)")
print("="*60)
print(f"- Sebelum: {text_clean}")
print(f"- Sesudah: {text_lower}")
print("\nPenjelasan: Semua huruf menjadi kecil untuk konsistensi pemrosesan")
```

---

```python
# 🔡 Langkah d: Tokenization
tokens = text_lower.split()

print("="*60)
print("🔡 LANGKAH d: TOKENIZATION (MEMISAHKAN TEKS MENJADI KATA-KATA)")
print("="*60)
print(f"- Teks: {text_lower}")
print(f"- Tokens: {tokens}")
print("\nPenjelasan: Memisahkan teks menjadi daftar kata (token) untuk diproses satu per satu")
```

---

```python
# 🔧 Langkah e: Normalization + Fuzzy Matching
print("="*60)
print("🔧 LANGKAH e: NORMALIZATION + FUZZY MATCHING")
print("="*60)

# 1. Dictionary Lookup
print("\n📚 1. Dictionary Lookup:")
print("-"*60)
for token in tokens:
    if token in NORMALIZATION_DICT.get("typo_corrections", {}):
        print(f"  '{token}' → TYPO → '{NORMALIZATION_DICT['typo_corrections'][token]}'")
    elif token in NORMALIZATION_DICT.get("abbreviations", {}):
        print(f"  '{token}' → SINGKATAN → '{NORMALIZATION_DICT['abbreviations'][token]}'")
    elif token in KAMUS_ALIAS:
        print(f"  '{token}' → KAMUS ALIAS → '{KAMUS_ALIAS[token]}'")
    else:
        print(f"  '{token}' → (tidak ditemukan di kamus langsung)")

# 2. Levenshtein Distance & Similarity Score
print("\n📐 2. Levenshtein Distance & Similarity Score:")
print("-"*60)

# Fungsi hitung Levenshtein Distance manual
def levenshtein(s1, s2):
    if len(s1) < len(s2):
        return levenshtein(s2, s1)
    if len(s2) == 0:
        return len(s1)
    previous_row = range(len(s2) + 1)
    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row
    return previous_row[-1]

# Coba fuzzy matching untuk setiap token
for token in tokens:
    best_match = None
    best_score = 0
    best_distance = 999
    
    if len(token) >=4 and not token.isdigit():
        for alias, canonical in KAMUS_ALIAS.items():
            distance = levenshtein(token, alias)
            max_len = max(len(token), len(alias))
            similarity = (max_len - distance) / max_len * 100
            
            if similarity > best_score and similarity >= 85:
                best_score = similarity
                best_match = canonical
                best_distance = distance
    
    if best_match:
        print(f"  '{token}' → FUZZY → '{best_match}' (Distance: {best_distance}, Similarity: {best_score:.1f}%)")
    else:
        print(f"  '{token}' → tetap")

# 3. Koreksi Kata
text_normalized = koreksi_teks(input_text)
print("\n✅ 3. Hasil Koreksi Kata:")
print("-"*60)
print(f"- Sebelum: {input_text}")
print(f"- Sesudah: {text_normalized}")
```

---

```python
# 📝 Langkah 4: Ekstraksi Entitas dengan Regex
print("="*60)
print("📝 LANGKAH 4: EKSTRAKSI ENTITAS DENGAN REGEX")
print("="*60)

hasil_ekstrak = ekstrak_entitas(text_normalized, teks_asli=input_text)
entitas = hasil_ekstrak["entitas"]

print("\nPola Regex yang digunakan:")
print("- Tanggal: r'(hari ini|kemarin|besok|tgl|tanggal|\\d{1,2}/\\d{1,2}(?:/\\d{2,4})?)'")
print("- Nama: r'(pak|bu|mas|mbak|kak|ibu|bapak)\\s+([a-z]+(?:\\s+[a-z]+)?)'")
print("- Jumlah: r'(\\d+(?:[.,]\\d+)?)\\s*(dus|karton|karto|box|bungkus|bks|pcs|unit|buah|renteng|pack|paket)'")
print("- Status: r'(lunas|hutang|belum bayar|blm bayar|cicil|dicicil|tempo)'")
print("- Metode: r'(tunai|cash|transfer|tf|trf|bank|qris)'")

print("\nHasil Ekstraksi:")
print("-"*60)
pprint.pprint(entitas)
```

---

```python
# 🎯 Langkah 5: Identifikasi Intent & Penyusunan Data Transaksi
print("="*60)
print("🎯 LANGKAH 5: IDENTIFIKASI INTENT & PENYUSUNAN DATA TRANSAKSI")
print("="*60)

results = proses_nlp(input_text)

print("\nIntent yang Diidentifikasi:")
print("-"*60)
for i, result in enumerate(results, 1):
    print(f"Intent {i}: {result['intent']}")

print("\nData Transaksi yang Disusun:")
print("-"*60)
for i, result in enumerate(results, 1):
    print(f"\nTransaksi #{i}:")
    pprint.pprint(result['entitas'])

print("\n" + "="*60)
print("✅ DATA SIAP DISIMPAN KE DATABASE!")
print("="*60)
```
