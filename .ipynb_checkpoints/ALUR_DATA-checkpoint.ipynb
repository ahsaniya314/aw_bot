# 📊 Alur Pemrosesan Data NLP (SUPER SIMPLE)

Notebook ini menunjukkan bagaimana data diproses dari awal sampai siap database!

---

```python
# Langkah 1: Import semua yang dibutuhkan
import sys
import os
sys.path.insert(0, os.path.abspath('.'))

from nlp.processor import proses_nlp
from nlp.normalizer import koreksi_teks
from nlp.extractor import ekstrak_entitas
from nlp.dictionaries import KAMUS_ALIAS

print("✅ Semua berhasil diimport!")
```

---

```python
# Langkah 2: Input Data (Data Mentah dari User)
input_text = "hari ini pak andi ambil willo 10 karto lunas dibayar tf"

print("="*60)
print("1. INPUT DATA")
print("="*60)
print(input_text)
```

---

```python
# Langkah 3: Text Cleaning & Preprocessing
import re

text_clean = re.sub(r'[^\w\s]', '', input_text)
text_clean = re.sub(r'\s+', ' ', text_clean).strip()
text_lower = text_clean.lower()
tokens = text_lower.split()

print("="*60)
print("2. PREPROCESSING")
print("="*60)
print(f"- Bersih: {text_clean}")
print(f"- Lowercase: {text_lower}")
print(f"- Token: {tokens}")
```

---

```python
# Langkah 4: Normalization & Fuzzy Matching (Koreksi Teks)
text_normalized = koreksi_teks(input_text)

print("="*60)
print("3. NORMALIZATION & FUZZY MATCHING")
print("="*60)
print(f"- Sebelum: {input_text}")
print(f"- Sesudah: {text_normalized}")
```

---

```python
# Langkah 5: Ekstraksi Entitas (dengan Regex)
hasil_ekstrak = ekstrak_entitas(text_normalized, teks_asli=input_text)
entitas = hasil_ekstrak["entitas"]

print("="*60)
print("4. EKSTRAKSI ENTITAS")
print("="*60)
import pprint
pprint.pprint(entitas)
```

---

```python
# Langkah 6: Proses Full & Hasil Akhir (Siap Database)
results = proses_nlp(input_text)

print("="*60)
print("5. HASIL AKHIR (SIAP DATABASE)")
print("="*60)
for i, result in enumerate(results, 1):
    print(f"\nTransaksi #{i}:")
    print(f"- Intent: {result['intent']}")
    pprint.pprint(result['entitas'])
```
