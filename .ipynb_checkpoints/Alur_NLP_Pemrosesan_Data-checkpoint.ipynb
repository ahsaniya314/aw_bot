# 📊 Alur Lengkap Pemrosesan Data NLP

Notebook ini menunjukkan **setiap langkah pemrosesan data** dari input teks mentah sampai menjadi data transaksi yang siap disimpan ke database!

---

```python
# Import semua module yang dibutuhkan
import sys
import os
from pprint import pprint

# Tambahkan path project agar bisa import module lokal
sys.path.insert(0, os.path.abspath('.'))

from nlp.processor import NLPProcessor
from nlp.normalizer import TextNormalizer
from nlp.extractor import EntityExtractor
from nlp.dictionaries import ITEM_ALIASES, SLANG_WORDS

print("✅ Semua modul berhasil diimport!")
```

---

# 1. INPUT DATA (Data Mentah dari User)

Data yang dikirim oleh pengguna melalui chat/teks.

```python
# ======================================
# LANGKAH a: INPUT DATA
# ======================================
input_text = "hari ini pak andi ambil willo 10 karto lunas dibayar tf"

print("=" * 60)
print("LANGKAH a: INPUT DATA")
print("=" * 60)
print(f"Input: {input_text}")
```

---

# 2. PREPROCESSING (Pembersihan & Persiapan Teks)

```python
normalizer = TextNormalizer()
extractor = EntityExtractor()
processor = NLPProcessor()

# ======================================
# LANGKAH b: TEXT CLEANING
# ======================================
import re
text_clean = re.sub(r'[^\w\s]', '', input_text)
text_clean = re.sub(r'\s+', ' ', text_clean).strip()

print("=" * 60)
print("LANGKAH b: TEXT CLEANING")
print("=" * 60)
print(f"Sebelum: {input_text}")
print(f"Sesudah: {text_clean}")
print()

# ======================================
# LANGKAH c: LOWERCASE
# ======================================
text_lower = text_clean.lower()

print("LANGKAH c: LOWERCASE")
print("=" * 60)
print(f"Sebelum: {text_clean}")
print(f"Sesudah: {text_lower}")
print()

# ======================================
# LANGKAH d: TOKENIZATION
# ======================================
tokens = text_lower.split()

print("LANGKAH d: TOKENIZATION")
print("=" * 60)
print(f"Teks: {text_lower}")
print(f"Tokens: {tokens}")
print()
```

---

# 3. NORMALIZATION & FUZZY MATCHING (Koreksi Teks)

```python
# ======================================
# LANGKAH e: NORMALIZATION + FUZZY MATCHING
# ======================================
print("LANGKAH e: NORMALIZATION + FUZZY MATCHING")
print("=" * 60)

# a. Dictionary Lookup + Levenshtein Distance + Similarity Score
print("\na. Dictionary Lookup + Levenshtein Distance + Similarity Score:")
print("-" * 60)
for token in tokens:
    if token in SLANG_WORDS:
        print(f"  '{token}' → SLANG → '{SLANG_WORDS[token]}'")
    elif token in ITEM_ALIASES:
        print(f"  '{token}' → ITEM ALIAS → '{ITEM_ALIASES[token]}'")
    else:
        # Coba fuzzy matching untuk item
        best_match = None
        best_score = 0
        for alias, canonical in ITEM_ALIASES.items():
            # Hitung Levenshtein Distance manual
            def levenshtein(s1, s2):
                if len(s1) < len(s2):
                    return levenshtein(s2, s1)
                if len(s2) == 0:
                    return len(s1)
                previous_row = range(len(s2) + 1)
                for i, c1 in enumerate(s1):
                    current_row = [i + 1]
                    for j, c2 in enumerate(s2):
                        insertions = previous_row[j + 1] + 1
                        deletions = current_row[j] + 1
                        substitutions = previous_row[j] + (c1 != c2)
                        current_row.append(min(insertions, deletions, substitutions))
                    previous_row = current_row
                return previous_row[-1]
            
            distance = levenshtein(token, alias)
            max_len = max(len(token), len(alias))
            similarity = (max_len - distance) / max_len * 100
            
            if similarity > best_score and similarity >= 85:
                best_score = similarity
                best_match = canonical
        
        if best_match:
            print(f"  '{token}' → FUZZY → '{best_match}' (similarity: {best_score:.1f}%)")
        else:
            print(f"  '{token}' → tetap")

# d. Koreksi Kata
text_normalized = normalizer.normalize(input_text)
print("\nd. Hasil Koreksi Kata:")
print("-" * 60)
print(f"Sebelum: {input_text}")
print(f"Sesudah: {text_normalized}")
```

---

# 4. EKSTRAKSI ENTITAS (Regex Pattern Matching)

```python
# ======================================
# EKSTRAKSI ENTITAS DENGAN REGEX
# ======================================
print("EKSTRAKSI ENTITAS DENGAN REGEX")
print("=" * 60)

entitas = extractor.extract(text_normalized)

print("\nPola Regex yang digunakan:")
print("- Tanggal: (hari ini|kemarin|lusa|\d{1,2}[-/\\s]\d{1,2}[-/\\s]\d{2,4})")
print("- Nama: (pak|bu|mas|mbak|kak|tante|om|paman|bapak|ibu)[\s]+([a-z]+(?:[\s]+[a-z]+)?)")
print("- Jumlah: (\d+(?:[.,]\d+)?)[\s]*(dus|karton|karto|box|bungkus|bks|pcs|unit|buah|buah|renteng|pack|paket)")
print("- Status: (lunas|hutang|belum bayar|blm bayar|cicil|dicicil|tempo)")
print("- Metode: (tunai|cash|transfer|trf|bank|qris|kredivo|shopee pay|spay|gopay|ovo|dana)")

print("\nHasil Ekstraksi:")
print("-" * 60)
pprint(entitas)
```

---

# 5. IDENTIFIKASI INTENT & PENYUSUNAN DATA TRANSAKSI

```python
# ======================================
# PROSES FULL DENGAN NLPProcessor
# ======================================
print("PROSES FULL DENGAN NLPProcessor")
print("=" * 60)

result = processor.process_message(input_text, user_id=123)

print("\nIntent yang Diidentifikasi:")
print("-" * 60)
print(f"Intent: {result['intent']}")

print("\nData Transaksi yang Disusun:")
print("-" * 60)
for i, trans in enumerate(result['transactions'], 1):
    print(f"\nTransaksi #{i}:")
    pprint(trans)

print("\n" + "=" * 60)
print("DATA SIAP DISIMPAN KE DATABASE!")
print("=" * 60)
```
